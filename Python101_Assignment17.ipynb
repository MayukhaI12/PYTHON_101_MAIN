```python
#@title 5.Word Frequency

'''
Write a program that reads the contents of a text file. 

Create a dictionary where the keys = individual words found in the file and values = number of times each word appears. 

Program should display the frequency of each word in a new file.
'''

# Create a text file with your own words (input.txt)
# Read the text file
# For each word you find, add it the dictionary (if it is not there in the dictionary)
## If it is there in the dictionary, simply update the COUNT
# at the end of the loop, print the KEYs (words) and the VALUE (count)

# Sample output
# Frequecy Table
# ant 4
# bat 3
# zebra 10
# snake 4

# Opening files

b = open("/content/HanselAndGretel.txt")
f = open("/content/x.txt",'w')

# Traversing the file "HanselAndGretel.txt" to write all the words into a list

c = b.read()
c = str(c)
ln = c.splitlines()
wd = []
for i in ln:
  for j in i.split(' '):
    wd.append(j)

# Finding the frequency of each word

ln = len(wd)
setw = set()
dictw = dict()
for i in range(ln):
  if wd[i] in setw:
    continue
  else:
    ct = 0
    ele = wd[i]
    setw.add(wd[i])
    for j in range(ln - i):
      if wd[j+i] == ele:
        ct = ct +1
    dictw[ele] = ct

# Printing the results and writing them in a new file (x.txt)

print("Frequency of each word in the story:")
print("------------------------------------")

for a in dictw:
  print(a, ":", dictw[a])
  f.write(str(a) + "  " + str(dictw[a]) + " " + "\n")
```
