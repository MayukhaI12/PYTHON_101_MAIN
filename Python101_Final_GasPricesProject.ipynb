```python
#@title All the global variables used:-

# prices_master_list and prices_mini_list are the global variables.

prices_master_list = []

# prices_master_list captures the file as a list of lists. 
# prices_mini_list is the name of the list(s) inside prices_master_list.
```

```python
#@title read_gas_prices()
'''
Method Name: read_gas_prices
Input: the file name (gasprices)
Returns: none
Global variables updated: prices_master_list

This function takes in the gasprices file name and creates a prices_master_list.
It then prints the prices_master_list and all the contents inside it.
'''

def read_gas_prices(file_name = 'gasprices.txt'):

  # Opening the 'gasprices' file.
  gasprices = open("/content/gasprices.txt")

  # Using a for loop to traverse through the file and organizing it into a prices_master_list and into prices_mini_list(s).
  for prices_mini_list in gasprices:
    prices_mini_list = prices_mini_list.rstrip()
    prices_mini_list = prices_mini_list.rsplit("-")
    prices_mini_list.append(prices_mini_list[-1].rsplit(":")[0])
    prices_mini_list.append(prices_mini_list[-2].rsplit(":")[1])
    prices_mini_list[-1] = prices_mini_list[-1].rstrip()
    prices_mini_list.pop(2)
    prices_master_list.append(prices_mini_list)

  # Writing the code to label all the elements inside a prices_mini_list.
    month = prices_mini_list[0]
    day = prices_mini_list[1]
    year = prices_mini_list[2]
    price = float(prices_mini_list[-1])

  # Writing the code to print the resuls.
  print("Printing: prices_master_list:-")
  print("------------------------------")
  print()

  for prices_mini_list in prices_master_list:
    print(prices_mini_list)

# ---------- #

# Calling the function.
read_gas_prices('gasprices.txt')
```

```python
#@title  get_average_price_per_year()

'''
Method Name: get_average_price_per_year
Input: None (It relies on the global variable prices_master_list.)
Returns: None
Global variables updated: none

This function finds the average gas price per year and prints the information in the following format:
Year      Average Price
1993     1.07
1994     1.09
'''

def get_average_price_per_year():

  # Setting up some variables to use later in the code.
  ct = 0
  sum = 0
  avg = []
  year = prices_master_list[0][2]

  # Using a for loop to traverse the prices_master_list and to find the answers.
  for i in prices_master_list:
    if(i[2] == year):
      ct = ct + 1
      sum = sum + float(i[3])
      year = i[2]
    else:
      avg.append([year, sum/ct])
      ct = 0
      sum = 0
      year = i[2]

  # Writing the code to print the results.
  print("Displaying average price per year:")
  print("-----------------------------------")

  print("Year ------ Average price")
  print()
  for i in avg:
    print(i[0], "------", i[-1])

# ---------- #

# Calling the function.
get_average_price_per_year()
```

```python
#@title get_average_price_per_month()

'''
Method Name: get_average_price_per_year
Input: None (It relies on the global variable prices_master_list.)
Returns: None
Global variables updated: none

This function finds the average gas price per month and prints it in the following format:
Month-Year      Average Price
04-1993         1.07
05-1993         1.09
'''

def get_average_price_per_month():

  # Setting up some variables to use later in the code. 

  count = 0
  sum = 0
  avg_month = []
  month = prices_master_list[0][0] + "-" + prices_master_list[0][2]

  # Traversing the prices_master_list using a for loop to find the average per month.
  for i in prices_master_list:
    curr_month = i[0] + "-" + i[2]
    if(curr_month == month):
      count = count + 1
      sum = sum + float(i[3])
      month = curr_month
    else:
      avg_month.append([month,sum/count])
      ct = 0
      sum = 0
      month = curr_month

  # Adjusting the code to print the results in the required format. (Month = Feb instead of Month = 02)

  # Printing the results.
  print("Displaying average price per month:")
  print("-----------------------------------")

  print("Month-Year ------ Average price")
  print()
  for i in avg_month:
    print(i[0], "------", i[-1])

    pass

# ---------- #

# Calling the function.
get_average_price_per_month()
```

```python
#@title get_highest_and_lowest_prices_per_year()

'''
Method Name: get_highest_and_lowest_prices_per_year
Input: None (It relies on the global variable prices_master_list.)
Returns: None
Global variables updated: none

This function finds the highest and lowest prices per year and prints it in the following format:

Displaying the highest price per year:

Highest price for the year 1993 : 1.068

---------------------------------

Displaying the lowest price per year:

Lowest price for the year 1993 : 1.068
'''

def get_highest_and_lowest_prices_per_year():

  # Setting up some variables for later use.
  list_of_years = []
  list_of_prices = []

  list_of_highest_prices = {}
  list_of_lowest_prices = {}

  x = []
  y = []

  # Using a for loop to process the prices_master_list to list all the years.
  for i in prices_master_list:
    i[-1] = float(i[-1])
    if (i[-2] not in list_of_years):
      list_of_years.append(i[-2])
    elif (i[-2] in list_of_years):
      continue

  # Processing the prices_master_list to find the highest and lowest prices per year.
  for i in prices_master_list:
    if (i[2] == list_of_years[0]):
      list_of_prices.append(i[-1])
      list_of_highest_prices.update({ i[-2] : max(list_of_prices)})

      list_of_lowest_prices.update({ i[-2] : min(list_of_prices)})

      list_of_prices.clear()
      list_of_years.pop(0)
      list_of_years.append(i[2])

  # Printing the results.
  print("Displaying the highest prices per year:")
  print()

  for a in list_of_highest_prices:
    print("Highest price for the year", a, ":", list_of_highest_prices[a])

  print()
  print("---------------------------------")
  print()

  print("Displaying the lowest prices per year:")
  print()

  for a in list_of_lowest_prices:
    print("Lowest price for the year", a, ":", list_of_lowest_prices[a])

    pass

# ---------- #

# Calling the function.
get_highest_and_lowest_prices_per_year()
```

```python
#@title generate_list_of_prices_lowest_to_highest()

'''
Method Name: generate_list_of_prices_lowest_to_highest
Input: None (It relies on the global variable prices_master_list.)
Returns: None
Global variables updated: none

This function sorts all the prices and prints them from lowest to highest.
'''

def generate_list_of_prices_lowest_to_highest():

  # Setting up an empty list beforehand to append to later.
  list_of_prices_lowest_to_highest = open('/content/list_of_prices_lowest_to_highest.txt','w')

  # Going through prices_master_list to sort all the prices.
  temp = sorted(prices_master_list, key=lambda x:x[3])

  # Using a for loop to print the results and write them into a file. 
  for i in range(0, len(temp)):
   print("[",temp[i][0],"-",temp[i][1],"-",temp[i][2],"]","---",temp[i][3])
   list_of_prices_lowest_to_highest.write("[" + str(temp[i][0]) + '-' + str(temp[i][1]) + "-" + str(temp[i][2]) +"]" + "---" + str(temp[i][3])+'\n')
    

  pass

# ---------- #

# Calling the function.
generate_list_of_prices_lowest_to_highest()
```

```python
#@title generate_list_of_prices_highest_to_lowest()

'''
Method Name: generate_list_of_prices_highest_to_lowest
Input: None (It relies on the global variable prices_master_list.)
Returns: None
Global variables updated: none

This function sorts all the prices and prints them from highest to lowest.
'''

def generate_list_of_prices_highest_to_lowest():

  # Opening an empty file to write the results.
  list_of_prices_highest_to_lowest = open('/content/list_of_prices_highest_to_lowest.txt','w')

  # Going through prices_master_list to sort all the prices.
  temp = sorted(prices_master_list, key=lambda x:x[3], reverse = True)

  # Using a for loop to print the results and write them into a file. 
  for i in range(0, len(temp)):
   print("[",temp[i][0],"-",temp[i][1],"-",temp[i][2],"]","---",temp[i][3])

   list_of_prices_highest_to_lowest.write("[" + str(temp[i][0]) + '-' + str(temp[i][1]) + "-" + str(temp[i][2]) +"]" + "---" + str(temp[i][3])+'\n')

  pass

# ---------- #

# Calling the function.
generate_list_of_prices_highest_to_lowest()
```
